```python
train_data = train_data.dropna(how = 'any')
```

```python
train_data.shape
```

```python
train_data['content'] = train_data['author']+" "+ train_data['title']
```

```python
train_data.head()
```

```python
from nltk.stem.porter import PorterStemmer
import nltk
nltk.download('stopwords')

#stemming
ps = PorterStemmer()

def stemming(content):
    stemmed_content = re.sub('[^a-zA-Z]'," ", content)
    stemmed_content = stemmed_content.lower()
    stemmed_content = stemmed_content.split()
    stemmed_content = [ps.stem(word) for word in stemmed_content if word not in stopwords.words('english')]
    stemmed_content = " ".join(stemmed_content)
    return stemmed_content
```

```python
train_data['content'] = train_data['content'].apply(stemming)
```

```python
x = train_data['content'].values
y = train_data['label'].values
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vector = TfidfVectorizer()
vector.fit(x)
x = vector.transform(x)
```

```python
print(x)
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 , random_state = 42)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()
rfc.fit(x_train , y_train)
```

```python
y_pred_rfc = rfc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_rfc = accuracy_score(y_test , y_pred_rfc)
accuracy_rfc
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()
dtc.fit(x_train , y_train)
```

```python
y_pred_dtc = dtc.predict(x_test)
```

```python
accuracy_dtc = accuracy_score(y_test , y_pred_dtc)
accuracy_dtc
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, GlobalMaxPooling1D, Dense

x = x.toarray().astype(str)
# Tokenize the text data
tokenizer = Tokenizer(num_words=1000)  # Adjust num_words based on vocabulary size
tokenizer.fit_on_texts(x)
sequences = tokenizer.texts_to_sequences(x)
X = pad_sequences(sequences, maxlen=20)  # Adjust maxlen based on input sequence length

# Define the CNN model
model = Sequential()
model.add(Embedding(input_dim=1000, output_dim=64, input_length=20))
model.add(Conv1D(128, 5, activation='relu'))
model.add(GlobalMaxPooling1D())
model.add(Dense(3, activation='softmax'))  # Adjust units based on the number of classes

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X, y, epochs=10, batch_size=32, validation_split=0.2)

# Evaluate the model
loss, accuracy = model.evaluate(X, y)
print("Accuracy:", accuracy)
```
