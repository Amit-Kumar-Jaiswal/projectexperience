```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.shodan.io/search?query=security+attacks+2023&page=1'
data = requests.get(url)
print(data)
```

```python
soup = BeautifulSoup(data.text , 'lxml')
print(soup)
```

```python
ipaddressess = []
ipaddress = soup.find_all('a' , class_ = 'title text-dark')
for i in ipaddress :
  ipadd = i.text
  ipaddressess.append(ipadd)
print(ipaddressess)
```

```python
companynames = []
cn = soup.find_all('a' , class_ = 'filter-link filter-org')
for i in cn :
  companyname = i.text
  companynames.append(companyname)
print(companynames)
```

```python
citynames = []
cityname = soup.find_all('a' , class_ = 'filter-link text-dark')
for i in cityname :
  cnames = i.text
  citynames.append(cnames)
print(citynames)

countrynames = []
cnames = []

for i in range(len(citynames)):
  if i % 2 == 0:  # Even index
    countrynames.append(citynames[i])
  else:  # Odd index
    cnames.append(citynames[i])

print(countrynames)
print(cnames)
```

```python
datetimes = []
datetime = soup.find_all('div' , class_ = 'timestamp u-pull-right text-secondary normalize')
for i in datetime :
  dt = i.text
  datetimes.append(dt)

print(datetimes)
```
