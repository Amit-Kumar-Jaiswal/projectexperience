```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://webscraper.io/test-sites/e-commerce/more'

data = requests.get(url)

print(data)
```

```python
soup = BeautifulSoup( data.text , 'lxml')
print (len(soup.text))
```

```python
itemboxes = soup.find_all('div' , class_ = 'card-body')
print (len(itemboxes))
```

```python
item_description = []

description = soup.find_all('p' , class_ = 'description card-text')

for i in description :
  descript = i.text
  item_description.append(descript)

print (item_description)
```

```python
item_prices = []

prices = soup.find_all('h4' , class_ = 'float-end price card-title pull-right')

for i in prices :
  price = i.text
  item_prices.append(price)

print (item_prices)
```

```python
item_reviews = []

reviews = soup.find_all('p' , class_ = 'float-end review-count')

for i in reviews :
  review = i.text
  item_reviews.append(review)

print(item_reviews)
```

```python
df = pd.DataFrame({'item_description' : item_description , 'item_prices' : item_prices , 'item_reviews' : item_reviews})
print(df)
```
