```python
firewall_value_counts = df2['Firewall Logs']

print(firewall_value_counts)

alerts_value_counts = df2['IDS/IPS Alerts'].value_counts()

print(alerts_value_counts)

log_source = df2['Log Source']

plt.ylabel('Counts')

plt.title('Distribution of Firewall logs in different log sources ( Firewall and Server)')

ax = sns.barplot(data = df2 , y = firewall_value_counts.index , x = firewall_value_counts , hue = log_source )

for i in ax.containers:
    ax.bar_label(i)
```
