```python
from gradientai import Gradient


def main():
    gradient = Gradient()

    base_model = gradient.get_base_model(base_model_slug="nous-hermes2")

    new_model_adapter = base_model.create_model_adapter(
        name="amit model"
    )
    print(f"Created model adapter with id {new_model_adapter.id}")

    new_model_adapter.fine_tune(samples=[{"inputs": "india, history, geography"}])
    #sample_query = "can you tell me a story set about India"
    sample_query = input("Please Enter what you want to search / your querry :")
    print(f"Asking: {sample_query}")

#Before fine tunning
    completion = new_model_adapter.complete(query=sample_query, max_generated_token_count=500).generated_output
    print(f"Generated: (before fine tunning) : {completion}")

    samples = [
        {"inputs": "### Instructions : About India"},
        {"inputs": "india is a highly populated country"}
    ]

    #Defining Parameters for fine tunning

    num_epoches = 3
    count = 0
    while count < num_epoches:
      print(f"Fine Tunning the model with iteration {count + 1}")
      new_model_adapter.fine_tune(samples = samples)
      count = count + 1

#After Fine tunning
      completion = new_model_adapter.complete(query=sample_query, max_generated_token_count=100).generated_output
      print(f"Generated: (After fine tunning) : {completion}")

    new_model_adapter.delete()
    gradient.close()

if __name__ == "__main__":
    main()
```
