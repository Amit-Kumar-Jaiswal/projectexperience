```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
titanic_train = pd.read_csv('/content/drive/MyDrive/Datasets/titanicdataset/train.csv')
titanic_test = pd.read_csv('/content/drive/MyDrive/Datasets/titanicdataset/test.csv')
```

```python
def clean(data) :
  data = titanic_train.drop(['PassengerId' , 'Name' , 'Cabin' , 'Ticket'] , axis = 1 , inplace = True)

  col = ['Age' , 'SibSp' , 	'Parch' , 'Fare']
  for i in col :
    data[i].fillna(data[i].mean() , inplace = True)

    data.Embarked.fillna('U' , inplace = True)
    return data

titanic_train = clean(titanic_train)
titanic_test = clean(titanic_test)
```
