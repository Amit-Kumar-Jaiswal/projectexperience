```python
from sklearn.utils import shuffle
df = shuffle(df)
```

```python
df.head()
```

```python
x = df.drop(['CLASS_LABEL'] , axis = 1)
y = df['CLASS_LABEL']
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 , random_state = 0)
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
x_train
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

lr.fit(x_train , y_train)
```

```python
x_train_predict = lr.predict(x_train)
x_train_predict
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.xlabel('predicted Value')
plt.ylabel('Actual Value')
plt.title('Distribution of predicted value and actual value of the training dataset')

sns.scatterplot(x = x_train_predict , y = y_train)
```

```python
sns.displot(x = x_train_predict - y_train , kind = 'kde')
```

```python
from sklearn.metrics import precision_score , precision_recall_curve , f1_score , recall_score , confusion_matrix
```

```python
score = confusion_matrix(x_train_predict , y_train)
print(score)
sns.heatmap(score)
```

```python
ps = precision_score(x_train_predict , y_train)
ps
```

```python
prc = precision_recall_curve(x_train_predict , y_train)
prc
```

```python
x_test_predict = lr.predict(x_test)
x_test_predict
```

```python
score_pred = confusion_matrix(x_test_predict , y_test)
print(score_pred)
sns.heatmap(score_pred)
```

```python
ps_test = precision_score(x_test_predict , y_test)
ps_test
```

```python
prc_test = precision_recall_curve(x_test_predict , y_test)
prc_test
```
