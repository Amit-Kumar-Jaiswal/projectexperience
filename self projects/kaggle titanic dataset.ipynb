# **Importing SimpleImputer module from Sklearn for imputing the integer values**



```python
from sklearn.impute import SimpleImputer
si = SimpleImputer(strategy='mean')
train_data['Age'] = si.fit_transform(train_data[['Age']])
test_data['Age'] = si.fit_transform(test_data[['Age']])
```

# **Importing SimpleImputer module from Sklearn for imputing the categorical values**

```python
sim = SimpleImputer(strategy='most_frequent')
train_data['Cabin'] = sim.fit_transform(train_data[['Cabin']])
train_data['Embarked'] = sim.fit_transform(train_data[['Embarked']])
test_data['Cabin'] = sim.fit_transform(test_data[['Cabin']])
test_data['Embarked'] = sim.fit_transform(test_data[['Embarked']])
```

# **Encoding the sex column using pandas Dummies**


```python
y_train_data = train_data['Survived']
features_for_encoding_dummies = ['Pclass', 'SibSp',
       'Parch' , 'Sex']
x_train_data = pd.get_dummies(train_data[features_for_encoding_dummies])
x_test_data = pd.get_dummies(test_data[features_for_encoding_dummies])
```

# **Importing random forest classifier with fine tune parameters**

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100, max_depth=6, random_state=42)
rfc.fit(x_train_data , y_train_data)
```

# **predicted results on test data**

```python
y_pred = rfc.predict(x_test_data)
```

# **Saved the results in the submission file**

```python
result = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': y_pred})
result.to_csv('submission.csv', index=False)
print("Successfully Saved")
```
