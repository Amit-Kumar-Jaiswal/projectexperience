```python
import numpy as np
```

```python
def get_user_input():
    num1 = float(input("Enter the first number : "))
    num2 = float(input("Enter the second numner : "))
    operation = input("Enter the operation sign (+, -, *, /) you want to perform : ")
    return num1, num2, operation
```

```python
def calculate(num1, num2, operation):
    if operation == "+" :
      result = np.add(num1, num2)
    elif operation == "-" :
      result = np.subtract(num1 , num2)
    elif operation == "*" :
      result = np.multiply(num1 , num2)
    elif operation == "/" :
       if num2 == 0 :
          print("Error : Number can not be divided. kindly Please enter higher number.")
          return
       else :
          result = np.divide(num1 , num2)
    else :
        print("Invalid Operation")
        return
    return result
```

```python
def display_result(result):
  print("This is your result : ", result)
```

```python
def main():
  num1, num2, operation = get_user_input()
  result = calculate(num1, num2, operation)
  if result is None :
    display_result(result)
```
