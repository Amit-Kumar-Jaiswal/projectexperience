```python
import numpy as np
```

```python
num1 = float(input("Enter the first number : "))
num2 = float(input("Enter the second numner : "))
operation = input("Enter the operation sign (+, -, *, /) you want to perform : ")
```

```python
if operation == "+" :
  result = np.add(num1, num2)
elif operation == "-" :
  result = np.subtract(num1 , num2)
elif operation == "*" :
  result = np.multiply(num1 , num2)
elif operation == "/" :
  if num2 == 0 :
    print("Error : Number can not be divided. kindly Please enter higher number.")
  else :
    result = np.divide(num1 , num2)
else :
  print("Invalid Operation")
```

```python
if "result" in locals():
  print("This is your result : ", result)
```
